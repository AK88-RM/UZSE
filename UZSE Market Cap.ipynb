```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\Alisher\Desktop\ddf.csv')
```

```python
df.head()
```

```python
df = df.drop('Unnamed: 0', axis=1)
df['Date'] = pd.to_datetime(df['Date'])
lastp = df.drop_duplicates(subset = 'ISIN', keep="last")
lastp.head()
```

```python
lastp.shape
```

```python
issdf = pd.DataFrame()  # create empty ndf
allURL = 'https://uzse.uz/isu_infos?begin_date=01.04.1994&end_date=02.02.2019&locale=en&page=%d'
for k in range(1, 5):
    url = allURL % k
    if pd.read_html(url)[0].empty:
        break
    else:
        chunk = pd.read_html(url)[0]
        issdf = pd.concat([issdf, chunk])

issdf.head()
```

```python
nissdf = issdf.drop([' # ', 'Logo', 'Current Price'], axis = 1)
nissdf.head()
```

```python
nissdf.dtypes
```

```python
cols = ['Category','Ticker','ISIN', 'Isu_Name', 'Listing_Date', 'Currency', 'Par_Value', 'Listed_Quantity']
nissdf.columns = cols
nissdf.head()
```

```python
nissdf['ISIN'] = nissdf['ISIN'] + ' ' + nissdf['Ticker']
nissdf.head()
```

```python
lastp = lastp.replace('  ',' ', regex=True)
lastp.head()
```

```python
nissdf['ISIN'] = nissdf['ISIN'].astype(str)
lastp['ISIN'] = lastp['ISIN'].astype(str)
```

```python
merged = pd.merge(nissdf, lastp[['ISIN', 'Security_Type', 'Trade_Price', 'Date', 'Time']], on = 'ISIN', how = 'left')
merged.head()
```

```python
merged['Market_Cap'] = np.where(merged['Trade_Price'].isnull(), 
                            merged.Listed_Quantity * merged.Par_Value, 
                            merged.Listed_Quantity * merged.Trade_Price)
merged.head()
```
